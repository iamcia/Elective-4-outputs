```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Check if the column contains string data
    if df[column].dtype == 'object':
        # Convert the text in the column to lowercase
        df[column] = df[column].str.lower()

print(df)
```

```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Check if the column contains string data
    if df[column].dtype == 'object':
        # Apply the regular expression substitution to each string in the column
        df[column] = df[column].apply(lambda x: re.sub("[^a-zA-Z0-9\s]",'', str(x)))

print(df)
```

```python
final = []
stop_words = stopwords.words('english')  # Get a list of English stop words

# Iterate over each cell in the DataFrame
for column in df.columns:
    for cell in df[column]:
        for word in str(cell).split():  # Convert cell to string before splitting
            if word.lower() not in stop_words:  # Check if word is a stop word (case-insensitive)
                final.append(word)

print(final)
```

```python
def remove_stopwords(article, stopwords): # This function takes a list of stopwords as an argument
    temp_fnl = []
    for word in str(article).split():
        if word.lower() not in stopwords:  # Use stop_words_list here
            temp_fnl.append(word)
    return temp_fnl

df['v2'] = df['v2'].apply(lambda x: remove_stopwords(x, stop_words))  # Pass the stop_words variable to the function
```

```python
df.head()
```

```python
# Tokenize the text in the 'v2' column and store the results in a new 'tokenized_text' column
df['tokenized_text'] = df['v2'].apply(lambda x: word_tokenize(' '.join(x)))
# Join the list of words back into a string
```

```python
# Convert the 'v2' column to string type if it's not already
df['v2'] = df['v2'].astype(str)

initial_metrics = pd.DataFrame({
    'Metric': ['Total Messages', 'Words per Message (Avg)', 'Unique Tokens', 'Stopword Removal (%)', 'Special Characters (%)'],
    'Before': [
        len(df),
        df['v2'].str.split().str.len().mean(),
        len(set(' '.join(df['v2']).split())),
        0,  # Placeholder for stopword removal percentage
        df['v2'].str.count('[^a-zA-Z0-9\s]').sum() / df['v2'].str.len().sum() * 100
    ]
})
```

```python
final_metrics = pd.DataFrame({
    'Metric': ['Total Messages', 'Words per Message (Avg)', 'Unique Tokens', 'Stopword Removal (%)', 'Special Characters (%)'],
    'After': [
        len(df),
        df['tokenized_text'].str.len().mean(),
        len(set([word for sublist in df['tokenized_text'] for word in sublist])),
        (initial_metrics.loc[initial_metrics['Metric'] == 'Words per Message (Avg)', 'Before'].values[0] -
         df['tokenized_text'].str.len().mean()) / initial_metrics.loc[initial_metrics['Metric'] == 'Words per Message (Avg)', 'Before'].values[0] * 100,
        0  # Special characters should be 0 after removal
    ]
})
```

```python
comparison = initial_metrics.merge(final_metrics, on='Metric')
display(comparison)
```
